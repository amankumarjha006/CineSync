```python
import numpy as np
import pandas as pd
import ast
```

__Loading files into DataFrames__

```python
movies_metadata = pd.read_csv("../data/raw/movies_metadata.csv", low_memory = False)
credits = pd.read_csv("../data/raw/credits.csv")
keywords = pd.read_csv("../data/raw/keywords.csv")
```

```python
movies_metadata.drop(columns=['adult', 'belongs_to_collection', 'homepage', 'imdb_id', 'production_countries', 'video', 'status', 'spoken_languages', 'tagline'], inplace = True)
```

```python
movies_metadata.columns
```

```python
movies_metadata = movies_metadata[movies_metadata['id'].apply(lambda x: str(x).isdigit())]
```

```python
movies_metadata['id'] = movies_metadata['id'].astype(int)
```

```python
movies_metadata = movies_metadata.drop_duplicates(subset='id')
credits = credits.drop_duplicates(subset='id')
keywords = keywords.drop_duplicates(subset='id')
```

__Merging the three Dataframes__

```python
movies = movies_metadata.merge(credits , on='id').merge(keywords, on='id')
```

```python
movies.shape
```

```python
movies = movies.dropna(subset=['overview', 'title'])
```

```python
movies.shape
```

```python
print(movies['genres'].iloc[0])
print(movies['cast'].iloc[0])
print(movies['keywords'].iloc[0])
```

```python
def parse_names(text):
    try:
        parsed = ast.literal_eval(text)
        return [item['name'] for item in parsed]
    except:
        return [] 
    
def get_top_cast(text):
    try:
        parsed = ast.literal_eval(text)
        return [item['name'] for item in parsed[:3]]
    except:
        return []
    
def get_director(text):
    try:
        parsed = ast.literal_eval(text)
        for item in parsed:
            if item['job'] == 'Director':
                return item['name']
        
    except:
        return " "
```

```python
movies['genres'] = movies['genres'].apply(parse_names)
movies['keywords'] = movies['keywords'].apply(parse_names)
movies['cast'] = movies['cast'].apply(get_top_cast)
movies['director'] = movies['crew'].apply(get_director)
```

```python
movies = movies.drop(columns=['crew'])
```

```python
movies[['title', 'genres', 'cast', 'director', 'keywords', 'overview']].head(3) 
```

```python
movies['release_year'] = pd.to_datetime(movies['release_date'], errors='coerce').dt.year
```

```python
movies['release_year'] = movies['release_year'].fillna(0).astype(int)
```

```python
movies.drop(columns=['release_date'], inplace=True)
```

```python
movies[['title', 'release_year']].head()
```

```python
def create_tags(row):
    genres = ' '.join(row['genres'])
    keywords = ' '.join(row['keywords'])
    cast = ' '.join(row['cast'])
    director = row['director']
    overview = row['overview']
    
    return genres + ' ' + keywords + ' '+  cast + ' ' + overview + ' '
```

```python
movies['tags'] = movies.apply(create_tags, axis=1)
```

```python
movies[['title', 'tags']].head()
```

```python
movies['tags'] = movies['tags'].str.lower().str.strip()
```

```python
movies[['title', 'tags']].head()
```

```python
movies.to_csv('../data/processed/movies_cleaned.csv', index=False)
```
